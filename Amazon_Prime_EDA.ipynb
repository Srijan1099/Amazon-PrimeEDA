```python
#check for null values
data.isnull().sum()
```

we have null values isn description column, So we are going to drop all the null values from description column

```python
#drop null calues in decription column
data.dropna(subset=['description'], inplace=True)
```

```python
#mode of age_certification
data['age_certification'].mode()[0]
```

```python
#replace all null values of age_certification with mode
data['age_certification'].fillna(data['age_certification'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

The 'seasons' column has null values because it's only applicable to TV shows, Movies don't have seasons so these entries would naturally be null, instead of removing these rows, replace it with 0.

```python
#replacing the null values in seasons column with 0
data['seasons'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

Since 'imdb_id' is unique for movies and tv shows we will drop all null entries in 'imdb_id' columns.

```python
data.dropna(subset=['imdb_id'], inplace=True)
```

```python
data.isnull().sum()
```

'imdb_score' represents the rating of the movies or TV  show, removing null calues in 'idb_score' might lead to data loss so, it's better to replace them with mean.

```python
round(data['imdb_score'].mean(),1)
```

# **Violin Plot**
Aviolin Plot is a statistical chart that combines aspects of a box plot and a kernel density plot to visualize the distributionof a numeric variable across different categories or groups.

```python
#violin plot of imdb score by type
plt.figure(figsize=(10,6))
sns.violinplot(data = data, x = 'type', y = 'imdb_score', color = 'orange')
plt.title('Violin plot of imdb score by type\n', color = 'brown')
plt.show()
```

**Insights**

. Movies: The distribution of imdb_score for movies shows that higher density of scores is between 6 and 7. This suggests that most
movies on Amazon Prime have decent ratings, with a tendency towards higher scores.

. TV Shows: The distribution of imdb_score for TV Shows that higher density of scores is between 7 and 8.

. The median imdb_score for Movies (indicated by the white dot inside the violin) is slightly lower than that of TV Shows. This suggests
that, on average, TV shows on Amazon Prime tend to have higher ratings than Movies.
